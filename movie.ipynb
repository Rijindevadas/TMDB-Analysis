```python
# Split the 'genres' column into multiple columns
genres_split= df2['genres'].str.split(', ', expand=True)

# Optionally rename the new columns
genres_split.columns=[f'genre_{i+1}' for i in range(genres_split.shape[1])]

# - genres_split.shape[1]: This gives the number of columns in the DataFrame.
# - range(genres_split.shape[1]): Creates a range of indices, like [0, 1, 2] if there are 3 columns.
# - f'genre_{i+1}': Uses Python f-strings to generate new names like 'genre_1', 'genre_2', etc.
# - The whole list comprehension builds a list of column names:


# Combine with the original DataFrame if needed
df2 = df2.drop(columns=['genres']).join(genres_split)
```

```python
df2
```

```python
df3=df2[['production_companies','production_countries','spoken_languages','keywords']]
df3
```

```python
df4=pd.concat([df2,df3],axis=1)
df4
```

```python
all_genres = pd.concat([df4[f'genre_{i+1}'] for i in range(genres_split.shape[1])])
genre_counts = all_genres.value_counts()

genre_counts.head(10).plot(kind='bar', figsize=(10, 5), color='purple')
plt.title("Top 10 Most Frequent Genres")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
genre_cols = [f'genre_{i+1}' for i in range(genres_split.shape[1])]
df_genres = df4[genre_cols + ['vote_average']].melt(id_vars='vote_average', value_name='genre').dropna()

genre_rating = df_genres.groupby('genre')['vote_average'].mean().sort_values(ascending=False)

# Plot
genre_rating.head(10).plot(kind='bar', figsize=(10,5), color='orange')
plt.title("Top 10 Genres by Average Rating")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
